```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import seaborn as sns

%matplotlib inline

datos = np.random.rand(100)

mpl.pyplot.hist(datos)
```

```python
sns.distplot(datos)
```

```python
sns.distplot(datos, color="green")
```

```python
datos2 = np.random.rand(80)
mpl.pyplot.hist(datos2, color="red")
```

```python
mpl.pyplot.hist(datos,color="green", bins=20 ,density=True)
mpl.pyplot.hist(datos2,color="blue", bins=20, density=True)
```

```python
datos3 = np.random.rand(100)
sns.jointplot(datos,datos3)
```

```python
sns.jointplot(datos,datos3, kind="hex")
```

```python
#Combinando estilos
datos5 = np.random.randn(100)

sns.distplot(datos5, color = "green", rug = False, hist=False)
```

```python
argumentos_curva = {'color':'black' , 'label': 'Curva'}
argumentos_histograma = {'color':'grey', 'label':'Histograma'}
sns.distplot(datos5, bins=25, kde_kws=argumentos_curva, hist_kws=argumentos_histograma)
```

```python
serie = pd.Series(datos5)
serie
```

```python
sns.distplot(serie, bins=25,color="blue")
```

```python
#regresiones
#entender como varia una variable respecto a otra
propinas = sns.load_dataset('tips')
propinas.head()
```

```python
sns.lmplot('total_bill','tip',propinas)
```

```python
sns.lmplot('total_bill','tip',propinas,scatter_kws={'marker':'o','color':'green'},line_kws={'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',propinas,fit_reg=False)
```

```python
propinas['porcentaje'] = 100 * propinas['tip']/propinas['total_bill']
```

```python
propinas.head
```

```python
sns.lmplot('size','porcentaje',propinas)
```

```python
sns.lmplot('total_bill','porcentaje',propinas,hue='sex',markers=['x','o'])
```

```python
sns.lmplot('total_bill','porcentaje',propinas, hue='day')
```

```python
#mapa de calor: Representacion grafica de datos, donde los valores individuales contenidos en una matriz se representan con colores
vuelos = sns.load_dataset('flights')
vuelos.head()
```

```python
vuelos = vuelos.pivot('month','year','passengers')
vuelos
```

```python
sns.heatmap(vuelos)
```

```python
sns.heatmap(vuelos,annot=True,fmt='d')
```

```python
valor_central = vuelos.loc['May'][1956]
valor_central
sns.heatmap(vuelos,center=valor_central, annot=True,fmt='d')
```

```python
sns.heatmap(vuelos,center=valor_central, annot=True,fmt='d',cbar_kws={'orientation':'horizontal'})
```
